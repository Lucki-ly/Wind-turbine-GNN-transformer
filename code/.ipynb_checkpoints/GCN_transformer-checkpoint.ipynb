```python
# Prepare numerical data
data_num = data[['风速', '功率', '风轮转速']]
X_std = StandardScaler().fit_transform(data_num.values)
data_std = pd.DataFrame(X_std)
```

```python
# Construct graph data using KNN
def build_knn_graph(data, k=5):
    adjacency_matrix = kneighbors_graph(data, k, mode='connectivity', include_self=True)
    edge_index = torch.tensor(np.array(adjacency_matrix.nonzero()), dtype=torch.long)
    return edge_index

edge_index = build_knn_graph(X_std)
data_tensor = torch.tensor(X_std, dtype=torch.float)
data_graph = Data(x=data_tensor, edge_index=edge_index)
```

```python
# Define GCN model
class GCN(torch.nn.Module):
    def __init__(self, input_dim, hidden_dim, output_dim):
        super(GCN, self).__init__()
        self.conv1 = GCNConv(input_dim, hidden_dim)
        self.conv2 = GCNConv(hidden_dim, output_dim)

    def forward(self, data):
        x, edge_index = data.x, data.edge_index
        x = self.conv1(x, edge_index)
        x = F.relu(x)
        x = self.conv2(x, edge_index)
        return x
# Train GCN model
def train_gcn(data_graph, epochs=100, lr=0.01):
    model = GCN(input_dim=data_graph.num_node_features, hidden_dim=16, output_dim=data_graph.num_node_features)
    optimizer = torch.optim.Adam(model.parameters(), lr=lr)
    for epoch in range(epochs):
        model.train()
        optimizer.zero_grad()
        output = model(data_graph)
        loss = F.mse_loss(output, data_graph.x)  # Ensure dimensions match
        loss.backward()
        optimizer.step()

        # Calculate additional metrics
        predictions = output.detach().numpy()
        ground_truth = data_graph.x.numpy()
        mse = mean_squared_error(ground_truth, predictions)
        mae = mean_absolute_error(ground_truth, predictions)

        # Binary classification for accuracy calculation
        binary_pred = (predictions >= np.mean(predictions, axis=0)).astype(int)
        binary_true = (ground_truth >= np.mean(ground_truth, axis=0)).astype(int)
        acc = np.mean([accuracy_score(binary_true[:, i], binary_pred[:, i]) for i in range(binary_true.shape[1])])

        if (epoch + 1) % 10 == 0:
            print(f"Epoch {epoch + 1}/{epochs}, Loss: {loss.item()}, Accuracy: {acc}, MSE: {mse}, MAE: {mae}")
    return model

model = train_gcn(data_graph)
data_graph.x = model(data_graph).detach()
```

```python
# Cluster with KMeans
def cluster_with_kmeans(data, n_clusters=10):
    kmeans = KMeans(n_clusters=n_clusters)
    data['cluster'] = kmeans.fit_predict(data)
    return data

clustered_data = cluster_with_kmeans(data_std, n_clusters=10)
```

```python
# Visualize anomalies
def visualize_anomalies(data_std, clustered_data, anomalies_col='anomaly'):
    plt.figure(figsize=(12, 8))

    # Scatter plot for anomalies
    anomaly_data = clustered_data[clustered_data[anomalies_col] == 1]
    normal_data = clustered_data[clustered_data[anomalies_col] == 0]

    plt.scatter(normal_data[0], normal_data[1], c='purple', label='Normal', alpha=0.6)
    plt.scatter(anomaly_data[0], anomaly_data[1], c='red', label='Anomaly', alpha=0.6)

    plt.title('Anomaly Detection Visualization')
    plt.xlabel('Standardized Feature 1')
    plt.ylabel('Standardized Feature 2')
    plt.legend()
    plt.grid(True)
    plt.show()

# Add anomaly detection
threshold = clustered_data[0].nlargest(int(0.4 * len(clustered_data))).min()
clustered_data['anomaly'] = (clustered_data[0] >= threshold).astype(int)

visualize_anomalies(data_std, clustered_data)
```

```python
import matplotlib.pyplot as plt

def plot_wind_speed_area(data):
    plt.figure(figsize=(14, 7))
    
    for wind_turbine in range(1, 13):
        turbine_data = data[data['风机编号'] == wind_turbine]
        plt.fill_between(turbine_data['时间戳'], turbine_data['风速'], alpha=0.3, label=f'风机编号 {wind_turbine}')
        plt.plot(turbine_data['时间戳'], turbine_data['风速'], alpha=0.6)  # 添加线条以增加可视化效果

    plt.title('风速与时间戳的关系')
    plt.xlabel('时间戳')
    plt.ylabel('风速')
    plt.legend()
    plt.grid()
    plt.show()

plot_wind_speed_area(data)
```

```python
def plot_predictions_vs_actual(data_graph, data_std):
    plt.figure(figsize=(14, 7))

    predicted_values = data_graph.x.numpy()
    actual_values = data_std.values

    plt.plot(predicted_values[:, 0], label='模型预测值', color='blue', alpha=0.6)
    plt.plot(actual_values[:, 0], label='实际值', color='pink', alpha=0.6)

    # Add threshold line
    threshold_value = np.mean(actual_values) + 2 * np.std(actual_values)  # Example threshold
    plt.axhline(y=threshold_value, color='red', linestyle='--', label='阈值线')

    plt.title('模型预测值与实际值对比')
    plt.xlabel('样本')
    plt.ylabel('风速')
    plt.legend()
    plt.grid()
    plt.show()
plot_predictions_vs_actual(data_graph, data_std)
```
